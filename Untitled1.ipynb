# Homework 6A 6.21)

* Partition theorem equals
$Z = \sum{e^{-E_{n}\beta}}$
* Where $E_n = \epsilon(1.03n - 0.03n^2)$ and $n = 1, 2, 3, ...$.
* To write $Z$ in terms of $KT/\epsilon$ we compute $Z=\sum e^{-(\frac{kT}{\epsilon})^{-1}(1.03n - 0.03n^2)}$
* Average Energy is $\bar E= \frac{E_n}{Z} e^{-(\frac{kT}{\epsilon})^{-1}(1.03n - 0.03n^2)}$
* Heat capacity is $C = \frac{1}{Z}\sum \frac{E_{n}^2}{kT^2} e^{-(\frac{kT}{\epsilon})^{-1}(1.03n - 0.03n^2)}$


### Part a)

```python
def E(n):
  return (1.03*n) - (0.03*n**2) # E_n without the epsilon
En5 = []
Z5no = []
Z5 = []
for n in range (5):
  En5.append(E(n)) # (1.03n - 0.03n^2)
  Z5no.append(exp(-En5[n]))  # e^(1.03n - 0.03n^2)
def Z_5(kTe):
  return sum(Z5no)*exp(1/kTe) # sum of e^(1.03n - 0.03n^2) * e^KT/epsilon
kTe_vals = [1,1.25,1.5,1.75,2,2.25,2.5,2.75,3]
for kTe in kTe_vals:
  Z5.append(Z_5(kTe))
```

```python
def E1(n):
  return (1.03*n) - (0.03*n**2)
En10 = []
Z10no = []
Z10 = []
for n in range (10):
  En10.append(E1(n))
  Z10no.append(exp(-En10[n]))
def Z_10(kTe):
  return sum(Z10no)*exp(1/kTe)
kTe_vals = [1,1.25,1.5,1.75,2,2.25,2.5,2.75,3]
for kTe in kTe_vals:
  Z10.append(Z_10(kTe))
```

```python
def E15(n):
  return (1.03*n) - (0.03*n**2)
En15 = []
Z15no = []
Z15 = []
for n in range (15):
  En15.append(E15(n))
  Z15no.append(exp(-En15[n]))
def Z_15(kTe):
  return sum(Z15no)*exp(1/kTe)
kTe_vals = [1,1.25,1.5,1.75,2,2.25,2.5,2.75,3]
for kTe in kTe_vals:
  Z15.append(Z_15(kTe))
```

```python
fig, ax = plt.subplots();
ax.plot(kTe_vals, Z5,label='Z_5', color='red');
ax.plot(kTe_vals, Z10,label='Z_10', color='yellow');
ax.plot(kTe_vals, Z15,label='Z_15', color='pink');
ax.set_xlabel("kTe");
ax.set_ylabel("Z");
ax.legend()
```
